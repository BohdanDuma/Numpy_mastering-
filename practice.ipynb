🏋️ Задача 1: Детекція аномалій (Advanced Indexing & Masking)У тебе є масив показників тиску в системі (100, 10) (100 датчиків, 10 вимірів).Дані: data = rng.uniform(50, 150, (100, 10)).Завдання: 1. Знайди середнє значення ($\mu$) та стандартне відхилення ($\sigma$) для всієї матриці.2. Створи маску для "аномалій" — значень, які виходять за межі $[\mu - 2\sigma, \mu + 2\sigma]$.3. Заміни всі аномалії на середнє значення ($\mu$) цього ж датчика (по рядку).DE-Challenge: Зроби заміну без жодного циклу, використовуючи np.where.

```python
rng = np.random.default_rng()
data = rng.uniform(50, 150, (100, 10))
mean = data.mean(axis=1, keepdims=True)
std = data.std(axis=1, keepdims=True)
mask = (data < (mean - 2*std)) | (data > (mean + 2*std))
data = np.where(mask, mean, data)
print(data)
```

### По стовпчику аксіс=0

```python
features = rng.integers(1, 1000, (50, 4))
means = features.mean(axis=0, keepdims = True)
stds = features.std(axis=0, keepdims = True)
data = (features - means)/stds
new_means = data.mean(axis=0, keepdims = True)
print(new_means)
```

У тебе є масив статус-кодів серверів codes = [200, 404, 500, 200, 301, 502, 200].Завдання: 1. Створи маску, яка вибирає лише "критичні помилки" (всі коди $\ge 500$).2. Використовуючи цю маску, виведи не самі коди, а їхні індекси (позиції у часі).3. Використовуючи отримані індекси, дістань відповідні значення з іншого масиву timestamps (просто створи масив такої ж довжини).DE-Challenge: Використай np.argwhere або np.where, щоб отримати індекси.

```python
codes = [200, 404, 500, 200, 301, 502, 200]
codes = np.array(codes)
mask = (codes >= 500)
np.argwhere(mask)
np.where(mask)[0]#Повертає плоский масив
```

```python
temps = rng.uniform(-10, 120, (5, 20))
```

```python
mask = (temps > 100)|(temps <= 0)
temps[mask] = np.nan 
print(temps)
medians = np.nanmedian(temps, axis=1, keepdims=True)

temps = np.where(mask, medians, temps)
temps
```
